```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('dialects_database.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
tables = cursor.fetchall()
```

```python
print("Tables in the database:")
for table in tables:
    print(table[0])
```

```python
df_text = pd.read_sql_query("SELECT * FROM id_text", conn)
df_text.head()
```

```python
df_dialect = pd.read_sql_query("SELECT * FROM id_dialect", conn)
df_dialect.head()
```

```python
print(df_text.shape)
print(df_dialect.shape)
```

```python
df = pd.merge(df_text, df_dialect, on='id')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('text_and_dialect.csv')
```
